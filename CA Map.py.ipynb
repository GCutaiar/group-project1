```python
#import dependencies 
import gmaps
import pandas as pd


# Google developer API key
from config import gkey

#Access maps with unique API key
gmaps.configure(api_key=gkey)
```

```python
#Import csv for diseases and place into a dataframe
path = 'Resource/vaccine-preventable-disease-cases-by-county-and-year-2.csv'
diseases= pd.read_csv(path)
diseases_pd = pd.DataFrame(diseases)

#Clean up data 
diseases_pd = diseases_pd[diseases_pd['year'] == 2017].drop(columns = ["year"])
diseases_pd.set_index("disease",inplace = True)
diseases_pd.drop(['Diphtheria','Hepatitis A','Hepatitis B, Acute','Hepatitis C, Acute',
                'Invasive Meningococcal Disease','Rubella','Varicella Hospitalizations'],
                inplace = True)

counties =['Los Angeles','Sonoma','Yuba','Alameda','Orange','Sacramento',   
            'San Diego','Kern','Ventura','San Mateo','Santa Barbara','Merced','El Dorado',
            'Napa','Sierra' ]

diseases_pd = diseases_pd[diseases_pd['county'].isin(counties)].reset_index()
diseases_pd =diseases_pd.groupby(['county']).sum()

diseases_pd
```

```python
#Create a list containing coordinates
coordinates = [
    (37.7799, -122.2822),
    (38.7426, -120.4358),
    (35.4937, -118.8597),
    (34.0522, -118.2437),
    (37.2010, -120.7120),
    
    (38.5025 , -122.2654),
    (33.7175 , -117.8311),
    (38.5816 , -121.4944),
    (32.7157 , -117.1611),
    (37.4337, -122.4014),
    
    (34.4208, - 119.6982),
    (39.5533 , -120.2513),
    (38.5780, - 122.9888),
    (34.3705, -119.1391),
    (39.2547 , -121.3999)]
```

```python
#Create map 
figure_layout = {
    'width': '550px',
    'height':'650px',
    'border':'1px solid black',
    'padding':'1px'
}
fig = gmaps.figure(layout = figure_layout)
```

```python
# Assign the marker layer to a variable
# markers = gmaps.marker_layer(coordinates)
locations = coordinates
weights = diseases_pd["count"]

#Add heatmap layer
heatmap_layer = (gmaps.heatmap_layer(locations, weights = weights))
fig.add_layer(heatmap_layer)

#Add intensity and radius of heatmap layer 
heatmap_layer.max_intensity = 200
heatmap_layer.point_radius = 20

fig

```
